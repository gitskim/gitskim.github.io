```python
import numpy as np
import scipy.ndimage as nd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def make_test_image(N=512):
    X = np.random.randint(0, 64, (N, N))

    for k in range(6):
        i0, j0 = np.random.randint(0, N-32, 2)
        i1 = np.random.randint(i0 + 8, N)
        j1 = np.random.randint(j0 + 8, N)
        X[i0 : i1, j0 : j1] = np.cumsum(
            np.random.randint(0, 6, (i1 - i0, j1 - j0)),
            axis=k%2
        )
    X = np.minimum(X, 255)
    return X
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(make_test_image())
```

```python
image = np.random.randint(0, 5, (16, 16))
```

```python
image = make_test_image()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(image)
```
